```r
library(ggplot2)
library(tibble)
library(pixmap)
library(matrixcalc)
```

```r
im <- read.pnm(file = "einstein.pgm")
print(im)
plot(im)
```

```r
Y.raw <- getChannels(im)
extract.blocks <- function(raw)
{
    col <- NULL
    
    for(i in seq(1, 512, 8))
    {
        for(j in seq(1, 512, 8))
        {
            col <- cbind(col, as.vector(raw[i:(i+7), j:(j+7)]))
        }
    }
    return(col)
}

Y <- extract.blocks(Y.raw)
dim(Y)
```

```r
random.orthogonal <- function(n=64)
{
    mat<-matrix(runif(n**2),nrow=n)/2
    return(mat+t(mat))
}

soft.threshold <- function(x, lam){sign(x)*pmax(abs(x)-lam, 0)}
l1.norm <- function(a) {sum(abs(a))}
metric <- function(D, Y){l1.norm(D%*%Y)}
loss.function <- function(D, Y, R, lam){frobenius.norm(D%*%R-Y)+lam*l1.norm(R)}
```

```r
ADM.step <- function(D, Y, lam)
{
    R <- soft.threshold(t(D)%*%Y, lam)
    svd.out <- svd(Y%*%t(R))
    D <- svd.out$u%*%t(svd.out$v)
    return(list(R=R, D=D))
}

```

```r
fit.ADM <- function(Y, D, lam=2, n_steps=1000)
{
    loss <- NULL
    metric <- NULL
    for(i in 1:n_steps)
    {
        ADM.update <- ADM.step(D, Y, lam)
        D <- ADM.update$D
        R <- ADM.update$R
        loss <- c(loss, loss.function(D, Y, R, lam))
        metric <- c(metric, metric(D, Y))
    }
    return(list(loss=loss, metric=rev(metric)[1]))
}
```

```r
metrics <- NULL
for(i in 1:50)
{
    D <- random.orthogonal()
    model <- fit.ADM(Y, D, n_steps=100)
    metrics <- c(metrics, model$metric)
}
```
